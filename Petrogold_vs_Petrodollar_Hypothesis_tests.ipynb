# **ASSESSING THE LEVEL OF FAMILIARITY WITH THE DEBATE OF PETRODOLLAR VS. PETROGOLD IN THE NIGRIAN OIL AND GAS INDUSTRY**

Combined familiarity of the debated concept as a 1D vector space may be obtained via a 2_D linear projection of familiarity with the individual concepts of petrodollar and petrogold to a 1D line using the principles of linear projections. For the two dimensions:
- Familiar with petrodollar
- Familiar with petrogold

A graduated scale of categorical selections **[ Very familiar, Somewhat familiar, Not familiar ]** is factorized into **[ 2, 1, 0 ]** respectively. The final distriution's mean is shifted by 1 to centralize the distribution around zero simply for semantic meaning of the null hypothesis.

Finally a one-sided and two-sided "one sample" T-test of the data is done against a population mean of zero (0) and a confidence level of 5% (0.05).

```python
def vectorize_column_ordered(column):
    """
    Vectorize a column with ordered categories.

    Args:
        column (pd.Series): The column to vectorize.

    Returns:
        pd.Series: The vectorized column.
    """

    # Extract unique categories and sort them alphabetically
    sorted_categories = sorted(column.unique())

    # Create a mapping of sorted categories to integer values
    category_mapping = {category: idx for idx, category in enumerate(sorted_categories)}

    # Map the original column to the integer values
    return column.map(category_mapping)


data['Petrogold familiarity'] = vectorize_column_ordered(data["6. How familiar are you with the concept of Petrogold?"])
data['Petrodollar familiarity'] = vectorize_column_ordered(data["7. How familiar are you with Nigeria's current petrodollar system?"])
data['familiarity'] = ( ( data['Petrogold familiarity'] + data['Petrodollar familiarity'] ) / 2 ) - 1

display(data)
```

```python
from matplotlib import pyplot as plt
data['familiarity'].plot(kind='hist', bins=20, title='familiarity')
plt.gca().spines[['top', 'right',]].set_visible(False)
```

```python
print(f"mean:{data['familiarity'].mean()}, standard deviation:{data['familiarity'].std()}")
```

```python
from scipy import stats

# Define the null and alternative hypotheses
H1 = "Being in the oil and gas industry drives the level of familiarity with the petrogold system and its petrodollar alternative"
H0 = "Being in the oil and gas industry has no significant effect on the level familiarity with the petrogold vs. petrodollar debate"
alpha = 0.025

pvalue = stats.ttest_1samp(a=data['familiarity'], popmean=0, alternative='two-sided')[1]

if pvalue < alpha:
    print(f"At p-value={pvalue} (vs. critical value={alpha})")
    print(f"    Reject the null hypothesis.\n    In favour of \"{H1}\"")
else:
    print(f"At {pvalue} (vs. critical value={alpha})")
    print(f"    Fail to reject the null hypothesis.\n    {H0}")
```

# **ASSESSING THE OPINION-BASED NET IMPACT OF A TRANSITION FROM PETRODOLLAR TO PETRGOLD**

The net impact of the transition from petrodollar to petrogold may be understood as a signifcance pairwise T-test comparison of the positive and negative contributors.

## **Positive contributors**
- Opinion-based potential economic benefits
- Strategic advantages obtainable
- Positive effects on operations in:
 - Contract negotiations
 - Export processes
 - Currency exchange stability
 - Investment flows
 - Financial reporting and accounting


## **Negative contributors**
- Opinion-based potential economic drawbacks
- Political risks and challenges
- Negative effects on operations in:
 - Contract negotiations
 - Export processes
 - Currency exchange stability
 - Investment flows
 - Financial reporting and accounting

## **Methodology/Steps**
- A probability score for each of the options to the multiple choice opinion responses is calculated. To do so, A weight is assigned to each option based on the frequency of selection as a ration of total choices.
- For each respondent a total choice score is calculated as the sum of the probability scores of all the selected options. By this method, selecting every option will yield a maximum score of 1 and no selections will yield 0.
- For responses on the effects on operation which have inherent positivity and negatvity built in a different approach is taken.
- Far all items a maximum score of 0.2 is assigned to the Very positive selection and -0.2 is assigned to the Very negative option. As susch a maximum score of 1 is obtainable for a selection of Very positive for all 5 scenarioss and a minimum score of -1 is obtanable for all selections as Very negative.
- The final net impact score is calculated as the sum of the positive contributors and the negatvie contributors as well as the unique question 10 in the following manner.<br><br>
<center>$Net\ impact=Positive\ contributor\ (Max:\ 1)-Negative\ contributor\ (Max:\ 1)+Net\ effect\ on\ operations$<center\><br><br>
- The net impact is then added as an extra column and plotted on a histogram.

```python
def prob_score(title):
    unique_keys = {}
    for item in data[title]:
        for key in item.split(";"):
            if key in unique_keys.keys():
                unique_keys[key] += 1
            else:
                unique_keys[key] = 1
    total = sum(unique_keys.values())
    for key in unique_keys.keys():
        unique_keys[key] = unique_keys[key] / total

    # unique_keys["total"] = sum(unique_keys.values())
    return unique_keys

def convert_column(title):
    column = prob_score(title)
    data[title] = [sum([column[key] for key in item.split(";")]) for item in data[title]]

for title in ['8. In your opinion, what would be the potential economic benefits of implementing a petrogold system in Nigeria? (Select all that apply)',
    '9. What are the potential economic drawbacks of implementing a petrogold system in Nigeria? (Select all that apply)',
    '11. What strategic advantages might Nigeria gain from adopting a petrogold system? (Select all that apply)',
    '12.What political risks or challenges might Nigeria face with the implementation of a petrogold system? (Select all that apply)']:
    convert_column(title)
display(data.head(2))
```

```python
cypher = {
    'Very positive':0.2,
    'Positive':0.1,
    'Neutral (Negligible impact)':0,
    'Negative':-0.1,
    'Very negative':-0.2
}

for col in ["Contract negotiations", "Export processes", "Currency exchange stability", "Investment flows", "Financial reporting and accounting"]:
    data[col] = data[col].map(cypher)
net_effect_on_operations = data[["Contract negotiations", "Export processes", "Currency exchange stability", "Investment flows", "Financial reporting and accounting"]].sum(axis=1)

data['Net impact'] \
= data['8. In your opinion, what would be the potential economic benefits of implementing a petrogold system in Nigeria? (Select all that apply)'] \
+ data['11. What strategic advantages might Nigeria gain from adopting a petrogold system? (Select all that apply)'] \
+ net_effect_on_operations \
- data['9. What are the potential economic drawbacks of implementing a petrogold system in Nigeria? (Select all that apply)'] \
- data['12.What political risks or challenges might Nigeria face with the implementation of a petrogold system? (Select all that apply)']

display(data)
```
